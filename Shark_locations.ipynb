```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import pearsonr, ttest_ind, spearmanr, t
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)


data_shark_url = "https://raw.githubusercontent.com/bekircan4721/Bekircan_arac-/main/attacks.csv"
data_temperature_url = "https://www.ncei.noaa.gov/data/oceans/woa/DATA_ANALYSIS/3M_HEAT_CONTENT/DATA/basin/pentad/pent_h22-w0-2000m.dat"

# reading datasets
data_shark = pd.read_csv(data_shark_url, encoding="unicode_escape")  # veya "ISO-8859-1"
data_temperature = pd.read_table(data_temperature_url, delim_whitespace=True)

# filtering by year
data_shark = data_shark[data_shark["Year"].notnull()]
print(data_shark.info())





# Counting attacks by location
location_attack = data_shark.groupby("Location").size().reset_index(name="Attack_Count")

# filtering the most 10 attack location
top_10_attack_locations = location_attack.sort_values(by="Attack_Count", ascending=False).head(10)



sns.barplot(x="Attack_Count", y="Location", data=top_10_attack_locations, palette="viridis")

plt.show()
```
